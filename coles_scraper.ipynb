```python
pages_dicitonary = dict(zip(categories, pages_per_category))
pages_dictionary
```

```python
products = []
brands = []
dollar_sale_price = []
product_save_price = []
cent_sale_price=[]
unit_size = []
product_category = []

for key, value in pages_dictionary.items():
    print(key, value)
    for file_number in range(1, value+1):
        with open(f"coles_scraping/{key}/{value}.html") as reader:
            html_string = reader.read()
            soup = BeautifulSoup(html_string, 'html.parser')

        for box in soup.find_all(class_="product product-specials"):
            try:
                product_save_price.append(box.find(class_="product-save-value").text[2:-2])
            except:
                product_save_price.append(0)
            products.append(box.find(class_="product-name").text)
            brands.append(box.find(class_="product-brand").text)
            dollar_sale_price.append(box.find(class_="dollar-value").text)
            cent_sale_price.append(box.find(class_="cent-value").text)
            unit_size.append(box.find(class_="package-size").text.replace(' on special', ''))
            product_category.append(key)
```

```python
data = pd.DataFrame(data={
    "brand": brands,
    "product": products,
    "dollar_sale_price": dollar_sale_price,
    "cent_sale_price": cent_sale_price,
    "saving": product_save_price,
    "packaging_size":unit_size,
    "product_category": product_category
})
data
```

```python
data['sale_price'] = data['dollar_sale_price']+data['cent_sale_price']
data.drop(axis=1, columns=["dollar_sale_price", "cent_sale_price"], inplace=True)
data
```
