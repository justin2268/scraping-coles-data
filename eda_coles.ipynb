```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_pickle("./coles_data.pkl")
data
```

```python
# get rid of products that have 0.0 as saving
data_dropped = data[data['saving'] != 0.0]
data_dropped
```

```python
count_of_products = data_dropped.groupby("product_category").count()
count_of_products
```

```python
# Count of the number of products in each Category
fig, ax = plt.subplots()

ax.barh(count_of_products.index, count_of_products['brand'])
plt.figure(figsize=(12,8))
ax.set_xticklabels(count_of_products['brand'], rotation=90)
plt.show()
```
